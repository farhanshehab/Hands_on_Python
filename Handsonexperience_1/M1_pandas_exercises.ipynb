```python
unique_course_count = df_courses.groupby('course_title')['subject'].nunique().reset_index()
```

```python
# Group by the 'subject' column and aggregate the 'course_title' as lists of unique course titles
courses_by_subject = df_courses.groupby('subject')['course_title'].unique().reset_index()

# Rename the columns for better readability
courses_by_subject.columns = ['subject', 'unique_course_titles']

# Display the result
print(courses_by_subject)
```

```python
# Group by the 'subject' column
grouped_courses = df_courses.groupby('subject')

# Get the unique course titles and their count for each subject
courses_with_count = grouped_courses.agg(
    unique_titles=('course_title', lambda x: list(x.unique())),
    title_count=('course_title', 'nunique')
).reset_index()

# Display the result
print(courses_with_count)

```

```python
# Use parentheses to call the groupby method and chain it with value_counts()
df_courses.groupby('subject').value_counts()
```
